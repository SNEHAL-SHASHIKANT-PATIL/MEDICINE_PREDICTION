```python
import numpy as np
import  pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds = pd.read_csv("Clean_Medicine_Prediction.csv")
```

```python
ds.info()
```

```python
ds.drop("Unnamed: 0" , axis = 1,inplace = True)
```

```python
sns.countplot( x = ds['Gender'])
plt.show()
```

```python
sns.boxplot(ds['Disease'])
plt.show()
```

```python
plt.figure(figsize = (10,8))
sns.countplot(x = ds['Disease'])
plt.xticks(rotation=45)
plt.show()
```

```python
top_disease = ds["Disease"].value_counts().head(5)
plt.figure(figsize = (10,6))
plt.pie(top_disease , autopct = '%1.2f%%')
plt.title("Top 5 Diseases ")
plt.show()
```
