```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
ds = pd.read_csv('medicine_prediction_dataset_650.csv')
```

```python
ds.info()
```

```python
ds['Age'].fillna(ds['Age'].median() , inplace = True)
```

```python
ds['Duration']  = ds['Duration'].replace(np.NaN ,'unknown Duration' )
```

```python
ds['Duration'].head(50)
```

```python
(ds['Duration'] == 'Months').sum()
```

```python
ds['Medicine Given'].isnull().sum()
```

```python
ds.dropna(subset=['Medicine Given'], inplace=True)
```

```python
ds['Age'] = ds['Age'].astype(int)
```

```python
ds.info()
```

```python
ds= ds.reset_index(drop = True)
```

```python
columns = ['Gender' , 'Symptoms' , 'Duration' , 'Disease' , 'Medicine Given']
for col in columns :
    ds[col] = ds[col].astype('string')
```

```python
ds.duplicated().sum()
ds.drop_duplicates(inplace  = True)
```

```python
ds.to_csv("Clean_Medicine_Prediction.csv")
```
